```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
o1a003_pred = pd.read_csv('purchase forecast datasets/outlet1_A003_predicted.csv')
o1a003_test = pd.read_csv('purchase forecast datasets/outlet1_A003_test.csv')
o1a003df = pd.DataFrame(o1a003_test)
o1a003df1 = pd.DataFrame(o1a003_pred['yhat'])
o1a003df['yhat'] = o1a003df1
o1a003df.head()

o1a003df = o1a003df.sort_values('ds')
o1a003df['ds'] = pd.to_datetime(o1a003df['ds'])

plt.plot(o1a003df['ds'], o1a003df['y'], 'r')
plt.plot(o1a003df['ds'], o1a003df['yhat'], 'b')
plt.show()
```
